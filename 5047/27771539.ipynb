```python
#BubbleSort for Divide and Conquer part
def sortByX(l): #input is list type
    size = len(l)
    new = []
    for i in l:
        new.append(i)
    
    for p_num in range(size-1,0,-1):
        for i in range(p_num):
            if new[i].getX() > new[i+1].getX():
                new[i], new[i+1] = new[i+1], new[i]
    return new

def sortByY(l): #input is list type
    size = len(l)
    new = []
    for i in l:
        new.append(i)
    
    for p_num in range(size-1,0,-1):
        for i in range(p_num):
            if new[i].getY() > new[i+1].getY():
                new[i], new[i+1] = new[i+1], new[i]
    return new
```

```python
#ADT Point Class
class Point:    
    #Constructor
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    #Setter
    def setValue(self, new_x, new_y):
        self.x = new_x
        self.y = new_y
        
    #Getter
    def getX(self):
        return self.x
    
    #Getter
    def getY(self):
        return self.y
    
    #Distence
    def dist(self, other_point):
        dx = abs(self.x - other_point.x)
        dy = abs(self.y - other_point.y)
        mDist = dx+dy
        return mDist
    
    #Print point
    def toPoint(self):
        return (self.x, self.y)
```

```python
#3a. Brute Force Implementation
def bruteForce(alist): 
    result = []
    if len(alist) < 2:
        print ("Need at least 2 points")
        return result
    else:
        size = len(alist)
        d = alist[0].dist(alist[1])
        result = [d, alist[0], alist[1]]
        for i, p in enumerate(alist):
            for j in range(i+1, size):
                if p.dist(alist[j]) < d:
                    d = p.dist(alist[j])
                    result = [d, p, alist[j]]
        return result
```

```python
#3b. Divide & Conquer Implementation
def DNC(Lx, Ly):
    result = []
    size = len(Lx)
    # when size <= 3, use bruteForce
    if size <= 3: 
        result = bruteForce(Lx)
        return result
    else:
        #Divid the Lx in half
        mid = size // 2 #Find the middle point in the sorted 
        mlineX = Lx[mid].getX()
        drawLine(mlineX)
        Lx_L = Lx[:mid]  #P[0] to P[n/2]
        Lx_R = Lx[mid:]  #P[n/2+1] to P[n]
        #Divid the Ly in half
        Ly_L = []
        Ly_R = []
        for i in Ly:
            if i in Lx_L:
                Ly_L.append(i)
            else:
                Ly_R.append(i)
        
        #Recursively call on closest_d()
        dL = DNC(Lx_L, Ly_L)
        dL_d = dL[0]
        dR = DNC(Lx_R, Ly_R)
        dR_d = dR[0]
        #Get delta
        d1 = min(dL_d, dR_d)
        pair = []
        if d1 == dL_d:
            result = dL
        else:
            result = dR
  
        #Find all candidate points located in the middle strip
        S = []
        for p in Ly:
            if abs(p.getX() - mlineX) <= d1:
                S.append(p)
        
        S_size = len(S)
        for i, p in enumerate(S):
            for j in range(i+1, i+16):
                if j < S_size:
                    if p.dist(S[j]) < d1:
                        d1 = p.dist(S[j]) #update the mini_d
                        result = [d1, p, S[j]]
        return result
```
