## 1. B

```python
import re
pattern = r'(?P<Date>\d{2}/\d{2}/\d{4}) (?P<Time>\d{2}:\d{2}:\d{1,2} [P|A]M)'
test = re.match(pattern, '28/01/2016 12:30:05 PM +0000')
test['Date']
```

```python
df_csv['Date'] = df_csv['UploadDate'].apply(lambda x:re.match(pattern,x)['Date'])
df_csv['Time'] = df_csv['UploadDate'].apply(lambda x:re.match(pattern,x)['Time'])
df_csv.drop('UploadDate',1,inplace=True)
```

```python
df_csv['lat'] = df_csv['Coordinates'].apply(lambda x: x.split(', ')[0])
df_csv['lon'] = df_csv['Coordinates'].apply(lambda x: x.split(', ')[1])
df_csv['lat'] = df_csv['lat'].apply(lambda x: x.lstrip('('))
df_csv['lon'] = df_csv['lon'].apply(lambda x: x.rstrip(')'))
df_csv.drop('Coordinates', 1, inplace = True)
```
