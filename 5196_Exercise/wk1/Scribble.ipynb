```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4],
                   [9, 8, np.nan, 6],
                  [np.nan, 2, np.nan, 10]],
                  columns=list('ABCD'))
df
```

```python
df.fillna(0)
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
values = {'A':'a',"B":'b',"C":'c',"D":'d'}
df.fillna(value = values)
```
