# Simulating random variables
R  by default has many distributions. For example, ?Normal tells us about four functions relating
to the normal distribution,
* __pnorm__ - the distribution function (cumulative distribution function (CDF))
* __rnorm__ - a function to generate normal random variates
* __qnorm__ - the quantile function (the inverse of the normal CDF )
* __dnorm__ - the density function
By adding {p,r,q,d} to {norm, pois, ubif, binom, exp, t, f, chisq} you can work with Normal, Poisson, Uniform, Binomial, Exponential, t, F, and Chi-square distrobutions. To get more information, do ?Poisson, ?Uniform, ?Binomial, ?TDist, ?FDist, ?chisquare.

For a continuous distribution (like the normal), the most useful functions for doing problems involving probability calculations are the "p" and "q" functions (cdf and inverse cdf), because the the density (pdf) calculated by the "d" function can only be used to calculate probabilities via integrals and R doesn't do integrals.

For a discrete distribution (like the binomial), the "d" function calculates the density (pmf), which in this case is a probability $P(X = x)$ and hence is useful in calculating probabilities.

# Binomial and Bernoulli Random variables

There are four different functions related to the Binomial distribution in R
* __pbinom__ - the distribution function (cumulative distribution function (CDF))
* __rbinom__ - a function to generate binomial random variates
* __qbinom__ - the quantile function (the inverse of the binomial CDF ).
The quantile is defined as the smallest value $x$ such that $F(x) \ge p$, where $F$ is the distribution function.
* __dbinom__ - the density function

# Continuous Uniform distribution
This is teh first continuous random variable. As you know, if $X$ a continuous random variable, $P(X=a)=0$. However, you might get dunif(a) a number that has nothing to do with the probbaility. For a continuous random variable, probabilities are calculated by the area under its PDF.

There are four different functions related to a continuous uniform distribution in R
* __punif__ - the distribution function (cumulative distribution function (CDF))
* __runif__ - a function to generate unifrom random variates
* __qunif__ - the quantile function (the inverse of the binomial CDF ).
The quantile is defined as the smallest value $x$ such that $F(x) \ge p$, where $F$ is the distribution function.
* __dunif__ - the density function


```r
x <- runif(1000) # a continuous uniform random variable takes values between 0 and 1
par(mfrow=c(1,2))
hist(x)
boxplot(x, horizontal=T)
par(old)
```

# Normal Distribution
* __pnorm__ :  is the R function that calculates the cdf $F(x) = P(X \le x)$ where $X$ is normal random variable.
* __qnorm__ : is the R function that calculates the inverse cdf, $F^{-1}$, of the normal distribution The cdf and the inverse cdf are related by $p=F(x)=P(X \le x)$ then $x=F^{-1}(x)$. So given a number $p$ between zero and one, qnorm looks up the $p^{th}$ quantile of the normal distribution.
* __dnorm__ : is the R function that calculates the pdf $f$ of the normal distribution. We can use this function to plot the normal density function.
* __rnorm__: the R function that simulates random variables having a specified normal distribution.

When you use rnorm(100), it is actually rnorm(100, mean=0, sd=1). If you want another mean and sd, you should specify them.

```r
pnorm(8, mean=10, sd=7) ## P(X <= 8) for X~N(10,, 7)
```

```r
# To plot a normal curve
x <- seq(-10, 30, 0.1)
y <- dnorm(x, 10, 7)
plot(x, y, type = "l", ylim=c(0, 0.06))
```

```r
qnorm(0.25, mean=4, sd=2)
```

```r
x <- seq(-3,10,length=200)
y <- dnorm(x,mean=4,sd=2)
plot(x,y,type="l")
x <- seq(-3,2.65,length=100)
y <- dnorm(x,mean=4,sd=2)
polygon(c(-3,x,2.65),c(0,y,0),col="red")
text(1,0.02,"0.25")
arrows(4,0.05,2.65,0,length=.15)
text(4,0.06,"2.65")
#http://rstudio-pubs-static.s3.amazonaws.com/78857_86c2403ca9c146ba8fcdcda79c3f4738.html
```

```r
# N(10, 3) is our population. We are going to sample from it.
z <- rnorm(200000, mean=10, sd=3) # a normal random variable of N(10, 3). 
hist(z,freq=FALSE,nclass=100)
x=seq(min(z),max(z),len=200)
lines(x,dnorm(x, mean=10, sd=3),col="red", lty=2, lwd=3) # True normal curve for N(10, 3)
lines(x,dnorm(x,mean=mean(z),sd=sqrt(var(z))), col="blue") # approximate normal curve using the sample
# As we have a lot sample point, they are extremely close.
```

We can find out what proportion of the deviates lie outside 3 standard deviations from the true mean, a common cutoff used by physical scientists. We can also see the true theoretical proportion:

```r
sum(abs((z-10)/3)>3)/length(z)
2*pnorm(-3)
```

```r
x <- rnorm(1000, mean=100, sd=15)
hist(x, probability=TRUE, , ylim=c(0, 0.030))
xx <- seq(min(x), max(x), length=100)
lines(xx, dnorm(xx, mean=100, sd=15), col="blue")
lines(density(x), col="red")
```

# Generic sampling
It is often useful to be able to randomly draw from a set of numbers or values we already have (for example when bootstrapping). This is done using the function __sample__. We supply sample with a vector of numbers or other values to draw from. Without any other arguments sample will randomly permute the vector. If in addition we supply an argument size, sample will randomly draw size numbers from our vector without replacement. There is also an argument replace that specifies whether drawing should be with replacement (by defualt without replacement). The argument prob can be used to specify a probability distribution other than uniform with which to draw the numbers.

```r
sample(10) # permutation
```

```r
sample(1:10)  # permutation
```

```r
sample(1:20, size=5) #randomly chooses five numbers between 1 and 20
```

```r
X <- c("Hi", "Hello", "Howdy", "Bye", "Oy")
sample(X, size=3)
```

```r
sample(X, size=4, replace=TRUE)
```

```r
sample(1:10, size = 15, replace = TRUE)
```

```r
# random sample of size 8 from sequence [5, 15]
set.seed(124)
sample(seq(5, 15), 8)
```

```r
# random sample of size 10 from sequence [1, 5] with unequal probabilities
# of being chosen
set.seed(124)
sample(5, 10, prob = c(0.3, 0.4, 0.1, 0.1, 0.1), replace = T)
```
