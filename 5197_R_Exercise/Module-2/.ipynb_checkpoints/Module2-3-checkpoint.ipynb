```r
str(cars)
```

This dataset has two numerical continuous variables. Let's draw a scatter plot of them.

```r
plot(cars) # scatter plot
```

Now, we add a regression line to see their linear relationship. We will cover regression later.

```r
plot(cars,
     main = "Speed vs. Stopping Distance for Cars",
     xlab = "Speed (MPH)",
     ylab = "Stopping Distance (feet)")
lin_fit <- lm(cars$dist ~ cars$speed)
abline(lin_fit, col = "red", lwd = 2) 
```

```r
# the correlation coefficient
cor(cars$speed, cars$dist)
```
